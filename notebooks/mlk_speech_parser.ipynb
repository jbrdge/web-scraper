```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://analytictech.com/mb021/mlk.htm"

page = requests.get(url)
soup = BeautifulSoup(page.text, "html")
```

```python
mlkj_speech = soup.find_all('p')
```

```python
speech_combined = [ p.text for p in mlkj_speech ]
string_speech = ' '.join(speech_combined)
string_speech = string_speech.replace('\r\n', ' ')
print(string_speech)
```

```python
import re
speech_punct_removed = re.sub(r'[^\w\s]', '', string_speech)
print(speech_punct_removed)
```

```python
speech_lower = speech_punct_removed.lower()
print(speech_lower)
```

```python
speech_broken_out = re.split(r'\s+', speech_lower)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(speech_broken_out, columns=['word'])
df_word_counts = df['word'].value_counts().reset_index()
print(df.head())
```

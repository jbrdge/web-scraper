```python
from bs4 import BeautifulSoup
import re
import requests
import pandas as pd
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html")
```

```python
table = soup.find('table', 'datatable w-full border border-zinc-200')
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
world_columns = table.find_all('th')

world_columns_titles = [col.get_text(strip=True).split('Â') for col in world_columns]
for i in range(len(world_columns_titles)):
    if len(world_columns_titles[i]) > 1:
        world_columns_titles[i] = world_columns_titles[i][0] + world_columns_titles[i][1]
    else: 
        world_columns_titles[i] = world_columns_titles[i][0]
    
print(world_columns_titles)
```

```python
df = pd.DataFrame(columns=world_columns_titles)
df

```

```python
table.find_all('td')
```

```python
table.find_all('tr')
```

```python
world_rows = table.find_all('tr')
for row in world_rows[1:]:
    data = row.find_all('td')
    row_data = [item.get_text(strip=True).replace('â\x88\x92','-') for item in data]
    print(row_data)
    df_length = len(df)
    df.loc[df_length] = row_data
```

```python
df.set_index('#', inplace=True)
df
```
